## Examining design matrices in Python

```python
import numpy as np
import statsmodels.api as sm
from patsy import dmatrices
```

Let's use the tips database.

```python
tips = sm.datasets.get_rdataset('tips','reshape2',cache=True).data
```

```python
tips.head()
```

Now we'll create and examine the design matrices using patsy (yes, I know it's deprecated, but I'll stop using it as soon as statsmodels does).

```python
y,X = dmatrices('tip ~ total_bill',data=tips)
```

```python
X
```

```python
y
```

Estimate the parameters using the classic matrix calculation:

$$\beta = (X^{T} X)^{-1} X^{T} y$$

```python
from numpy.linalg import inv
β = inv(X.T @ X) @ X.T @ y
β
```

Calculate the fitted values:

$$\hat{y} = X\beta$$

```python
ŷ = X @ β
ŷ
```

Now we'll compute the residual or error term,  which, as usual, is the difference between the original y values and the fitted values:
$$e = y - \hat{y}$$

```python
e = y - ŷ
```

The original y values are the fitted values plus the error term,:

$$ y = \hat{y} + e = Xβ + e $$

```python
    X @ β + e
```
