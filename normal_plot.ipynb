Plot of a normal distribution pdf.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
%matplotlib inline
```

Set up an x axis with NumPy and plot with Matplotlib. We'll be using the Seaborn library to set the theme, even though we won't use any of its plotting functions.

```python
sns.set_theme()
x = np.linspace(-3,3)
plt.plot(x,norm.pdf(x))
```
