## Investigating the normal distribution with SymPy and calculus

```python
from sympy import *
init_printing()
x,μ,σ = symbols('x μ σ')
%matplotlib inline
```

```python
f = 1 / (sqrt(2*pi)*σ) * exp(Rational(-1,2)*((x - μ) / σ)**2)
```

```python
f
```

```python
f = f.subs([(μ,0),(σ,1)])
```

```python
f
```

```python
plot(f,(x,-4,4))
```

There's the classic bell curve.

```python
f_prime = diff(f,x)
```

```python
f_prime
```

```python
f_prime.subs(x,2)
```

```python
f_prime.subs(x,-2)
```

Due to the symmetry of the curve, slopes on opposite sides are mirror images of each other with different signs.

```python
integrate(f,x)
```

```python
N(integrate(f,(x,-2,2)))
```

Integrating f is the same as using the classic normal tables, but using scipy.stats.norm is probably more practical.

```python
f.subs(x,2)
```

```python
N(_)
```

```python
f.subs(x,0)
```

```python
N(_)
```

```python
N(integrate(f,(x,-1,1)))
```

```python
N(integrate(f,(x,-0.5,1)))
```

```python
f_prime.subs(x,0.5)
```

```python
N(_)
```

```python
f_prime.subs(x,-0.5)
```

x < 0 is positively sloped, and x > 0 is negatively sloped.

The maximum of this curve is $x = 0$.

Let's see if we can maximize it with the more formal test.

```python
solve(f_prime,x)
```

```python
diff(f,x,x)
```

```python
solve(_,x)
```

```python
f_double_prime = diff(f,x,x)
```

```python
f_double_prime.subs(x,0)
```

With the second-derivative test < 0, f is maximized at x = 0. Let's try looking at the curve symbolically again.
